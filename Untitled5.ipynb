```python
def transfer_masked_pattern(source_img, mask_img, target_img, src_points, dst_points, lower_color, upper_color):
    # マスク画像から特定の色を持つ領域を抽出
    mask_colored = cv2.inRange(mask_img, lower_color, upper_color)
    
    # 変換行列を取得
    matrix = cv2.getPerspectiveTransform(src_points, dst_points)
    
    # perspective変換を適用して領域を抽出
    warped_img = cv2.warpPerspective(source_img, matrix, (target_img.shape[1], target_img.shape[0]))
    warped_mask = cv2.warpPerspective(mask_colored, matrix, (target_img.shape[1], target_img.shape[0]))
    
    # dst_points からバイナリマスクを作成
    rect_mask = np.zeros((target_img.shape[0], target_img.shape[1]), dtype=np.uint8)
    dst_points_int = np.int32(dst_points.reshape(-1, 1, 2))
    cv2.fillConvexPoly(rect_mask, dst_points_int, 255)
    
    # はみ出し部分をクリップ
    warped_mask = cv2.bitwise_and(warped_mask, rect_mask)
    
    # 3チャネルのマスク画像を作成します
    warped_mask_3channel = cv2.merge([warped_mask, warped_mask, warped_mask])
    
    # マスクから領域を抽出
    masked_pattern = cv2.bitwise_and(warped_img, warped_img, mask=warped_mask)
    
    # target_imgに抽出した領域を重ねる
    result = cv2.bitwise_and(target_img, cv2.bitwise_not(warped_mask_3channel))
    result = cv2.add(result, masked_pattern)

    return result
```
