```python
import glob
import os.path
import numpy as np
from PIL import Image
import tensorflow as tf
import csv


FLAGS = tf.compat.v1.flags.FLAGS

tf.compat.v1.flags.DEFINE_string('original_gt_folder',
                                 './VOCdevkit/VOC2012/SegmentationClass',
                                 'Original ground truth annotations.')

tf.compat.v1.flags.DEFINE_string('segmentation_format', 'png', 'Segmentation format.')

tf.compat.v1.flags.DEFINE_string('output_dir',
                                 './VOCdevkit/VOC2012/SegmentationClassRaw',
                                 'folder to save modified ground truth annotations.')
tf.compat.v1.flags.DEFINE_string('class_dict_csv',
                                 './class_dict.csv',
                                 'Path to the CSV containing color to class id mapping.')


# def _remove_colormap(filename):
#   """Removes the color map from the annotation.

#   Args:
#     filename: Ground truth annotation filename.

#   Returns:
#     Annotation without color map.
#   """
#   return np.array(Image.open(filename))

def load_colormap_and_save_fixed(class_dict_csv, output_csv_path):
    """CSVファイルからカラーマッピングを読み込み、修正したマッピングを保存する。

    Args:
        class_dict_csv: カラーとクラスIDのマッピングが記述されたCSVファイルのパス。
        output_csv_path: 修正されたカラーマッピングを保存するCSVファイルのパス。

    Returns:
        カラーマッピングの辞書。
    """
    colormap_to_classid = {}
    class_name_to_id = {}
    current_id = 0

    with open(class_dict_csv, mode='r') as csvfile, open(output_csv_path, mode='w', newline='') as outputfile:
        reader = csv.DictReader(csvfile)
        fieldnames = ['r', 'g', 'b', 'class_id', 'name']  # Add 'class_id' and 'name' to the output
        writer = csv.DictWriter(outputfile, fieldnames=fieldnames)
        writer.writeheader()

        for row in reader:
            r, g, b, class_name = int(row['r']), int(row['g']), int(row['b']), row['name']
            if class_name not in class_name_to_id:
                class_name_to_id[class_name] = current_id
                current_id += 1

            class_id = class_name_to_id[class_name]
            colormap_to_classid[(r, g, b)] = class_id

            # Write the new mapping to the output CSV
            writer.writerow({'r': r, 'g': g, 'b': b, 'class_id': class_id, 'name': class_name})

    return colormap_to_classid

def _remove_colormap(filename, colormap_to_classid):
    """画像からカラーマップを除去し、クラスIDに変換する。

    Args:
        filename: セグメンテーションアノテーションファイル名。
        colormap_to_classid: カラーからクラスIDへのマッピング。

    Returns:
        クラスIDに変換されたアノテーション。
    """
    image = Image.open(filename)
    image_np = np.array(image)

    class_id_map = np.zeros(image_np.shape[:2], dtype=np.uint8)  # 画像の高さと幅に合わせて初期化

    for (r, g, b), class_id in colormap_to_classid.items():
        indices = (image_np[:, :, 0] == r) & (image_np[:, :, 1] == g) & (image_np[:, :, 2] == b)
        class_id_map[indices] = class_id

    return class_id_map


def _save_annotation(annotation, filename):
  """Saves the annotation as png file.

  Args:
    annotation: Segmentation annotation.
    filename: Output filename.
  """
  pil_image = Image.fromarray(annotation.astype(dtype=np.uint8))
  with tf.io.gfile.GFile(filename, mode='w') as f:
    print(filename)
    pil_image.save(f, 'PNG')


def main(unused_argv):
  # Create the output directory if not exists.
  if not tf.io.gfile.isdir(FLAGS.output_dir):
    tf.io.gfile.makedirs(FLAGS.output_dir)

  output_csv_path = os.path.join(os.path.dirname(FLAGS.class_dict_csv), 'class_dict_fix.csv')
  colormap_to_classid = load_colormap_and_save_fixed(FLAGS.class_dict_csv, output_csv_path)


  annotations = glob.glob(os.path.join(FLAGS.original_gt_folder,
                                       '*.' + FLAGS.segmentation_format))
  for annotation in annotations:
    raw_annotation = _remove_colormap(annotation, colormap_to_classid)
    filename = os.path.basename(annotation)[:-4]
    _save_annotation(raw_annotation,
                     os.path.join(
                         FLAGS.output_dir,
                         filename + '.' + FLAGS.segmentation_format))


if __name__ == '__main__':
  tf.compat.v1.app.run()
```

```python
from PIL import Image
import os
from tqdm import tqdm

# train.txtのファイルパス
train_txt_path = 'test.txt'

# 出力ディレクトリの作成
output_dirs = ['JPEGImages', 'SegmentationClass']
for output_dir in output_dirs:
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

# train.txtを読み込み、各行を処理
with open(train_txt_path, 'r') as file:
    for line in tqdm(file):
        # スペースで分割して画像パスとアノテーションパスを取得
        img_path, annot_path = line.strip().split(' ')
        
        # ファイル名の取得
        img_filename = os.path.basename(img_path)
        annot_filename = os.path.basename(annot_path)
        
        # 画像の読み込みとリサイズ
        img = Image.open("./"+os.path.basename(os.path.dirname(img_path))+"/"+os.path.basename(img_path))
        annot = Image.open("./"+os.path.basename(os.path.dirname(annot_path))+"/"+os.path.basename(annot_path)[:-4]+"_c.png")
        img_resized = img.resize((480, 240))
        annot_resized = annot.resize((480, 240))
        
        # リサイズした画像の保存
        img_resized.save(os.path.join('JPEGImages', img_filename))
        annot_resized.save(os.path.join('SegmentationClass', img_filename))

# ImageSets/Segmentationディレクトリの作成
output_dir = 'ImageSets/Segmentation'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 新しいtrain.txtファイルのパス
new_train_txt_path = os.path.join(output_dir, train_txt_path)

# train.txtを読み込み、各行から画像ファイル名を抽出して新しいtrain.txtに保存
with open(train_txt_path, 'r') as file, open(new_train_txt_path, 'w') as new_file:
    for line in file:
        # スペースで分割して画像パスを取得
        img_path = line.strip().split(' ')[0]
        
        # ファイル名（拡張子除く）の取得
        img_filename_without_ext = os.path.splitext(os.path.basename(img_path))[0]
        
        # 新しいtrain.txtにファイル名を書き込む
        new_file.write(img_filename_without_ext + '\n')

print("ImageSets/Segmentation/train.txtにファイル名を保存しました。")
print("画像とアノテーションのリサイズ及び保存が完了しました。")
```
